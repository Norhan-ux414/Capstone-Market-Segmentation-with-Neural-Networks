```python
feature_cols = [
    "age", "job", "marital", "education",
    "housing", "loan", "contact", "month", "day_of_week",
    "duration", "campaign", "pdays", "previous", "poutcome",
    "emp.var.rate", "cons.price.idx", "cons.conf.idx",
    "euribor3m", "nr.employed"
]

X= df[feature_cols].copy()
X.head()
```

```python
numeric_cols = [
    "age", "duration", "campaign", "pdays", "previous",
    "emp.var.rate", "cons.price.idx", "cons.conf.idx",
    "euribor3m", "nr.employed"
]

categorical_cols = [
    "job", "marital", "education",
    "housing", "loan", "contact",
    "month", "day_of_week", "poutcome"
]

print("Numeric columns:", numeric_cols)
print("Categorical columns:", categorical_cols)
```

```python
X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

X_encoded.head()
X_encoded.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = X_encoded.copy()

X_scaled[numeric_cols] = scaler.fit_transform(X_scaled[numeric_cols])

X_scaled.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val = train_test_split(
    X_scaled,
    test_size=0.2,
    random_state=42,
    shuffle=True
)

X_train.shape, X_val.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

input_dim = X_train.shape[1]   


input_layer = keras.Input(shape=(input_dim,))


encoded = layers.Dense(32, activation="relu")(input_layer)
encoded = layers.Dense(16, activation="relu")(encoded)   # bottleneck (Ø§Ù„ØªÙ…Ø«ÙŠÙ„ Ø§Ù„Ù…Ø¶ØºÙˆØ·)


decoded = layers.Dense(32, activation="relu")(encoded)
decoded = layers.Dense(input_dim, activation="linear")(decoded)


autoencoder = keras.Model(inputs=input_layer, outputs=decoded)

autoencoder.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss="mse"
)

autoencoder.summary()
```

```python
history = autoencoder.fit(
    X_train,
    X_train,         
    epochs=30,
    batch_size=256,
    shuffle=True,
    validation_data=(X_val, X_val),
    verbose=1
)
```

```python
# Ù†Ø­Ø¯Ø¯ Ø·Ø¨Ù‚Ø© Ø§Ù„Ù€ bottleneck (dense_1 Ø­Ø³Ø¨ Ø§Ù„Ù€ summary)
encoder = keras.Model(
    inputs=autoencoder.input,
    outputs=autoencoder.get_layer(index=2).output  # layer Ø±Ù‚Ù… 2 Ù‡ÙŠ (None, 16)
)

# Ø§Ø³ØªØ®Ø±Ø§Ø¬ Ø§Ù„ØªÙ…Ø«ÙŠÙ„ Ø§Ù„Ù…Ø¶ØºÙˆØ· Ù„ÙƒÙ„ Ø§Ù„Ø¹Ù…Ù„Ø§Ø¡
X_encoded_latent = encoder.predict(X_scaled)

X_encoded_latent.shape
```

```python
import pandas as pd

latent_cols = [f"z{i}" for i in range(1, 17)]
latent_df = pd.DataFrame(X_encoded_latent, columns=latent_cols)

latent_df.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X_latent = latent_df.values  

inertias = []
k_values = range(2, 11)   
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_latent)
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(6,4))
plt.plot(k_values, inertias, marker="o")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia (Within-Cluster SSE)")
plt.title("Elbow Method on Latent Space")
plt.show()
```
